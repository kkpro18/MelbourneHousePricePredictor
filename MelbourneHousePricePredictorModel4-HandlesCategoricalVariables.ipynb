```python
import pandas as pd
import numpy as np
import sklearn
```

```python
melbourne_data = pd.read_csv("melb_data.csv")
melbourne_data.head()
```

```python
y = melbourne_data.Price # our target - what we try to predict
x = melbourne_data.drop(['Price'], axis=1) # all features except price (our predictor)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=18)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
def finetune_return_mae(x_train, x_test, y_train, y_test):
    finetuned_model = RandomForestRegressor(n_estimators=100, random_state=18)
    finetuned_model.fit(x_train, y_train)
    y_predicted_values = finetuned_model.predict(x_test)
    mae = mean_absolute_error(y_test, y_predicted_values)
    return mae
```

```python
columns_with_missing_values = []
for column in x_train.columns:
    if x_train[column].isnull().any():
        columns_with_missing_values.append(column)
x_train.drop(columns_with_missing_values, axis=1, inplace=True)
x_test.drop(columns_with_missing_values, axis=1, inplace=True)
```

## Handling Categorical variables

```python
# cardinality means number of unique values in a column
low_cardinality_columns = []
for column in x_train.columns:
    if x_train[column].nunique() < 10 and x_train[column].dtype == "object":
        low_cardinality_columns.append(column)
# adds columns with unique number values less than 10
```

```python
numerical_columns = []
for column in x_train.columns:
    if x_train[column].dtype in ['float64', 'int64']:
        numerical_columns.append(column)
```

```python
# combine only these two types of data in our data
data = low_cardinality_columns + numerical_columns
x_train = x_train[data].copy()
x_test = x_test[data].copy()
```

```python
x_train.head()
```

```python
# Get the list of these categorical variables
categorical_variables_names = (x_train.dtypes == 'object') # selects the other data type
categorical_variables_names = list(categorical_variables_names[categorical_variables_names].index)
categorical_variables_names
```

Approach 1: Drop the categorical columns if they are not useful

```python
x_train_drop_categorical_columns = x_train.select_dtypes(exclude=['object'])
x_test_drop_categorical_columns = x_test.select_dtypes(exclude=['object'])
print("Approach 1: Dropping these categorical columns")
finetune_return_mae(x_train_drop_categorical_columns, x_test_drop_categorical_columns, y_train, y_test)
```

Approach 2: Ordinal Encoding - Assigns each value to a different integer, with ordered values e.g worst, bad, ok, good, best - 0, 1, 2, 3, 4, 5. Works best for tree-based models like random forest and decision tree

```python
from sklearn.preprocessing import OrdinalEncoder

OE_x_train = x_train.copy()
OE_x_test = x_test.copy()
OE = OrdinalEncoder()

OE_x_train[categorical_variables_names] = OE.fit_transform(OE_x_train[categorical_variables_names])
OE_x_test[categorical_variables_names] = OE.transform(OE_x_test[categorical_variables_names])

print("Approach 2: Ordinal Encoding")
finetune_return_mae(OE_x_train, OE_x_test, y_train, y_test)
```

Approach 3: One-hot Encoding - Creates new columns that indicates presence or absence of these variables e.g column for variable red, with 0/1 for presence/absence and additional columns for other colors. It works best for nominal variables such as colors, which values have no ranking

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

OHE_x_train = pd.DataFrame(OHE.fit_transform(x_train[categorical_variables_names]))
OHE_x_test = pd.DataFrame(OHE.transform(x_test[categorical_variables_names]))

OHE_x_train.index = x_train.index
OHE_x_test.index = x_test.index

numerical_x_train = x_train.drop(categorical_variables_names, axis=1)
numerical_x_test = x_test.drop(categorical_variables_names, axis=1)

OHE_x_train = pd.concat([OHE_x_train, numerical_x_train], axis=1)
OHE_x_test = pd.concat([OHE_x_test, numerical_x_test], axis=1)

OHE_x_train.columns = OHE_x_train.columns.astype(str)
OHE_x_test.columns = OHE_x_test.columns.astype(str)

print("Approach 3: OneHotEncoder")
finetune_return_mae(OHE_x_train,OHE_x_test,y_train,y_test)
```

The best approach is Approach 2 using Ordinal Encoding. The worst was the first approach; removing the columns with categorical data. Approach 2 and Approach 3 had very similar MAE scores.

```python
from sklearn.ensemble import RandomForestRegressor
melbourne_model = RandomForestRegressor(n_estimators=100, random_state=18) # random state is like seeding, to permit reproducibility
# n_estimators=100
melbourne_model.fit(OE_x_train,y_train) # fitting model with data
```

```python
# Here, we predict house price using the train and test data on the model
y_train_pred = melbourne_model.predict(OE_x_train)
y_test_pred = melbourne_model.predict(OE_x_test)
```

```python
from sklearn.metrics import r2_score
# here we can assess the r2 score of the model to check reliability and accuracy of the model, against exposed and unexposed data
r2_score_train = r2_score(y_train, y_train_pred)
r2_score_test = r2_score(y_test, y_test_pred)
```
